```python
#visualization shows the more/least sales according to oultet type
Outlet_Type_pivot= train_data.pivot_table(index='Outlet_Type', values='Item_Outlet_Sales', aggfunc=np.median)

Outlet_Type_pivot.plot(kind='bar', figsize=(10,10))
plt.xlabel('Outlet_Type')
plt.ylabel('Item_Outlet_Sales')
plt.title('Impact of Outlet Type on Item Oultet Sales')
plt.xticks(rotation=0)
#plt.show()
```

```python
#visualization shows the more/least sales according to fat content show there is no impact
Item_Fat_Content_pivot= train_data.pivot_table(index='Item_Fat_Content', values='Item_Outlet_Sales', aggfunc=np.median)

Item_Fat_Content_pivot.plot(kind='bar', figsize=(10,10))
plt.xlabel('Item_Fat_Content')
plt.ylabel('Item_Outlet_Sales')
plt.title('Impact of Fat Content on Item Oultet Sales')
plt.xticks(rotation=0)
#plt.show()
```

```python
#count item fat content
train_data['Item_Fat_Content'].value_counts()
```

```python
#replacing the repeated names
#train_data = train_data['Item_Fat_Content'].replace({'LF':'Low Fat','reg':'Regular', 'low fat':'Low Fat'})
train_data['Item_Fat_Content'] = train_data['Item_Fat_Content'].replace({'LF':'Low Fat','reg':'Regular','low fat':'Low Fat',})
```

```python
#count item fat content
train_data['Item_Fat_Content'].value_counts()
```

```python
#visualization shows the more/least sales according to fat content show there is no impact after replacing
Item_Fat_Content_pivot= train_data.pivot_table(index='Item_Fat_Content', values='Item_Outlet_Sales', aggfunc=np.median)

Item_Fat_Content_pivot.plot(kind='bar', figsize=(10,10))
plt.xlabel('Item_Fat_Content')
plt.ylabel('Item_Outlet_Sales')
plt.title('Impact of Fat Content on Item Oultet Sales')
plt.xticks(rotation=0)
#plt.show()
```

```python
#count missing values
missing_values= train_data.isnull().sum()
missing_values
```

```python
#total missing values
total_missing_values= missing_values.sum()
print("total missing values= ", total_missing_values)

#count total cells
total_cells = np.product(train_data.shape)
print("total number of cells= ", total_cells)

#percentage of total missing values
total_missing_values_percentage= (total_missing_values/total_cells)*100
print("percent of total missing values= %" ,total_missing_values_percentage)
```

```python
#since it is normal destribution we will Replace null values with mean
train_data['Item_Weight'].mean()
```

```python
#missing values have been replaced with the mean using the fillna function.
train_data['Item_Weight'].fillna(train_data['Item_Weight'].mean(), inplace=True) 
```

```python
train_data['Outlet_Size'].value_counts()
```

```python
#since the most are medium and the data is not recorded we can assume the null values as medium (mean)
train_data['Outlet_Size'].fillna('Medium', inplace=True)
```

```python
#no null values remained
train_data.isnull().sum()
```

```python
#when visibility is 0 then it means that the product was not available which means place empty 
train_data[train_data['Item_Visibility']==0]['Item_Visibility'].count()
```

```python
#since visibility graph is right skew then its better to take the median to fill the NaN
train_data['Item_Visibility'].fillna(train_data['Item_Visibility'].median())
```

```python
#new column contain Changing only the first 2 characters idetifier (i.e. the category ID)

train_data['New_Item_type'] = train_data['Item_Identifier'].apply(lambda x: x[0:2])
```
